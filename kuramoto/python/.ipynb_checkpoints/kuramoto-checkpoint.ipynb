```python
import numpy as np
import numpy.random as rd
import matplotlib.pyplot as plt
import functools as ft
import concurrent.futures
import scipy.integrate as integrate
import scipy.optimize as optimize
```

```python
def theta_deriv_general(thetas, t, K, N, omegas):
#     return omegas + (K/N) * np.sum(np.sin(thetas-thetas[:,None]), axis=0)
    return omegas + (K/N) * ( np.sum(np.sin(thetas)) * np.cos(thetas) -  np.sin(thetas) * np.sum(np.cos(thetas)))

def evolveAndStore(deriv, x0, t0, dt, T):    
    x = x0
    ts = np.linspace(t0, T, int((T-t0+dt)/dt))
    store = np.empty([ts.size, x.size])
    store[0] = x
    
    for i, t in enumerate(ts[:-1]):
        x += dt * deriv(x, t)
        store[i+1] = x
    
    return ts, store

N = 10
dK = 0.5
K  = [0,5]
Ks = np.linspace(K[0], K[1], int((K[1]-K[0]+dK)/dK))

fig, axs = plt.subplots(nrows=len(Ks), ncols=1, figsize = (10, 5), constrained_layout=True)
axs = axs.ravel()

Thetas00 = rd.uniform(0, 2*np.pi, N)
Omegas00 = rd.normal(size = N)

for i,k in enumerate(Ks):
    # Global definitions
    Thetas0 = np.copy(Thetas00)
    Omegas = np.copy(Omegas00)
    
    theta_deriv = ft.partial(theta_deriv_general, K=k, N=N, omegas=Omegas)
    ts, matrix = evolveAndStore(theta_deriv, x0 = Thetas0, t0 = 0, dt = 0.01, T = 100)
    
    for line in matrix.transpose():
#         axs[i].set_title('K = {}'.format(round(k,2)))
#         axs[i].set_ylabel(r'sin$\theta$')
#         axs[i].set_xlabel('t')
        axs[i].axis('off')
        axs[i].plot(ts, np.sin(line))
        
# fig.add_subplot(111, frameon=False)
# # hide tick and tick label of the big axis
# plt.tick_params(labelcolor='none', top=False, bottom=False, left=False, right=False)
# plt.xlabel(r'$\sin\left(\theta_{i}\right)$')
# plt.ylabel("common Y")

fig.text(0.5, 0, r'$t$', ha='center')
fig.text(0, 0.5, r'$\sin\left(\theta_{i}\right)$', va='center', rotation='vertical')
plt.savefig('../lyx/images/theta-normal.png', format='png', dpi=500)
```

```python
def r_infinity_K(N, ts, dt, K, thetas, omegas):
    
    for t in ts[:-1]:
        thetas += dt * (omegas + (K/N) * ( np.sum(np.sin(thetas)) * np.cos(thetas) 
                                         -np.sin(thetas) * np.sum(np.cos(thetas))))
    
    return np.abs(np.sum(np.exp(1j*thetas)))

def r_infinity_ks_paralell(N, T, dt, K, dk, thetas, omegas):
    t0, k0 = 0, 0
    ts = np.linspace(t0, T, int((T-t0+dt)/dt))
    ks = np.linspace(k0, K, int((K-k0+dk)/dk))
    rs = np.empty(ks.size)
    
    with concurrent.futures.ProcessPoolExecutor() as executor:
        
        results = [executor.submit(r_infinity_K, N, ts, dt, k, thetas, omegas) for k in ks]

        for i, f in enumerate(concurrent.futures.as_completed(results)):
            rs[i] = f.result()/N
            
    return ks, rs

def r_infinity_ks(N, T, dt, K, dk, thetas1, omegas1):
    thetas2 = np.copy(thetas1)
    omegas2 = np.copy(omegas1)
    t0, k0 = 0, 0
    ts = np.linspace(t0, T, int((T-t0+dt)/dt))
    ks = np.linspace(k0, K, int((K-k0+dk)/dk))
    rs = np.empty(ks.size)
    
    for i,k in enumerate(ks):
        rs[i] = r_infinity_K(N, ts, dt, k, thetas2, omegas2)
            
    return ks, rs/N
```

```python
def gaussian(w):
    return np.exp(-w**2/2) / np.sqrt(2*np.pi)

def integrand_(K, r, theta):
    return K * np.cos(theta)**2 * gaussian(K*r*np.sin(theta))

N = 1000
for i in range(5):
    Thetas = rd.uniform(0, 2*np.pi, N)
    Omegas = rd.normal(size = N)
    ks, rs = r_infinity_ks(N, 100, 0.01, 5, 0.2, Thetas, Omegas)
    plt.plot(ks, rs, lw = 0.1, color = 'k')
    plt.plot(ks, rs, '.')

k0, K, dk = 0, 5, 0.2
ks = np.linspace(k0, K, int((K-k0+dk)/dk))
rs_test = np.linspace(0, 1, 100)
rs_solution = np.empty(ks.size)

for i,k in enumerate(ks):
    is_test = np.empty(rs_test.size) 
    for j,r in enumerate(rs_test):
        integrand = ft.partial(integrand_, k, r)
        is_test[j] = np.abs(integrate.quad(integrand, -np.pi/2, np.pi/2)[0] - 1)
    rs_solution[i] = rs_test[np.argmin(is_test)]

plt.plot(ks, rs_solution, 'k-', label = 'theoretical')
        
plt.ylabel(r'$r_{+\infty}$')
plt.xlabel(r'$K$')
plt.legend()
plt.savefig('../lyx/images/rk1.png', format='png', dpi=500)
```

```python
def r_instant(N, T, dt, K, tthetas, oomegas):
    thetas = np.copy(tthetas)
    omegas = np.copy(oomegas)
    t0 = 0
    ts = np.linspace(t0, T, int((T-t0+dt)/dt))
    
    store = np.empty([ts.size, thetas.size])
    store[0] = thetas
    
    for i,t in enumerate(ts[:-1]):
        thetas += dt * (omegas + (K/N) * ( np.sum(np.sin(thetas)) * np.cos(thetas) 
                                         -np.sin(thetas) * np.sum(np.cos(thetas))))
        store[i+1] = thetas
    
    return ts, np.abs(np.sum(np.exp(1j*store), axis=1))/N
```

```python
ks = [1, 2]
plt.ylabel(r'$r(t)$')
plt.xlabel(r'$t$')

N = 1000
tthetas = rd.uniform(0, 2*np.pi, N)
oomegas = rd.normal(size = N)

for k in ks:
    ts, r = r_instant(N, 100, 0.01, k, tthetas, oomegas)
    plt.plot(ts, r, label = 'K = {}'.format(k))
plt.legend()
plt.savefig('../lyx/images/rt2.png', format='png', dpi=500)
```

```python
def uniform(w):
    if abs(w) <= 0.5:
        return 1
    else:
        return 0

def integrand_uni(K, r, theta):
    return K * np.cos(theta)**2 * uniform(K*r*np.sin(theta))

N = 2000
for i in range(5):
    Thetas = rd.uniform(0, 2*np.pi, N)
    Omegas = rd.uniform(-0.5, 0.5, N)
    ks, rs = r_infinity_ks(N, 200, 0.05, 1.5, 0.03, Thetas, Omegas)
    plt.plot(ks, rs, lw = 0.1, color = 'k')
    plt.plot(ks, rs, '.')

k0, K, dk = 0, 1.5, 0.03
ks = np.linspace(k0, K, int((K-k0+dk)/dk))
rs_test = np.linspace(0, 1, 100)
rs_solution = np.empty(ks.size)

for i,k in enumerate(ks):
    is_test = np.empty(rs_test.size) 
    for j,r in enumerate(rs_test):
        integrand = ft.partial(integrand_uni, k, r)
        is_test[j] = np.abs(integrate.quad(integrand, -np.pi/2, np.pi/2)[0] - 1)
    rs_solution[i] = rs_test[np.argmin(is_test)]

plt.plot(ks, rs_solution, 'k-', label = 'theoretical')
        
plt.ylabel(r'$r_{+\infty}$')
plt.xlabel(r'$K$')
plt.legend()
plt.savefig('../lyx/images/rk3.png', format='png', dpi=500)
```
